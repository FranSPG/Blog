```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange
```

```python
def graph(x, y, title):
    plt.figure(figsize=(25, 14))
    plt.scatter(x, y, s=np.pi*3)
    plt.ticklabel_format(useOffset=False, style='plain')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.savefig(title + '.jpg')
    plt.show()
```

```python
n = 1000
m = 2
x = list(range(1, n))
```

```python
def power(n, m):
    return np.power(n, m)
```

```python
y_power_m = [power(num, m) for num in x]
graph(x, y_power_m, 'potencia de 2')
```

```python
def square_root(n):
    return np.sqrt(n)
```

```python
n = 1000
x = list(range(1, n))
y_square_root = [round(square_root(num), 0) for num in x]
```

```python
graph(x, y_square_root, 'raiz cuadrada, redondeado a 0 dgitos')
```

```python
def digits(n):
    
    if n < 10:
        return [n]

    return digits(n//10) + [n % 10]
```

```python
def mult_digits(n):

    return n - np.prod([x for x in digits(n) if x != 0])    
```

```python
n = 1000
x = list(range(1, n))
y_mult_digits = [mult_digits(num) for num in x]
```

```python
graph(x, y_mult_digits, 'resta de la multiplicacion de los digitos de un numero (distintos de 0)')
```

```python
def fly_straight_dammit(n, memo = {}):
    
    if n <= 1:
        return 1
    
    try:
        n_1 = memo[n-1]
        
    except:
        n_1 = fly_straight_dammit(n-1)
        if np.gcd(n_1, n) > 1:
            return (n_1 // np.gcd(n_1, n))
    
        return n_1 + n + 1
```

```python
#n = 1000
#x = list(range(1, n))
y_fly_straight_dammit = [fly_straight_dammit(num) for num in x]
```

```python
graph(x, y_fly_straight_dammit, 'fly straight dammit')
```

```python
def rest_reverse_prime(n):
    return n - int(format(n, 'b')[::-1], 2)
```

```python
n = 300000
x = list(primerange(1, n))
y_rest_reverse_prime = [rest_reverse_prime(num) for num in x]
```

```python
graph(x, y_rest_reverse_prime, 'resta del numero binario invertido')
```
